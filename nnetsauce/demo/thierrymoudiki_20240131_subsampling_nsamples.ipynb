```python
import nnetsauce as ns
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(1968801)
size = 1000
mu_x = 100
sigma_x = 25
#x = np.random.normal(mu_x, sigma_x, size=size)
x = np.random.beta(a = 0.1, b = 0.2, size = size)
```

```python
w = 0
subx = ns.SubSampler(y = x, row_sample=0.3, seed = 1)
idx = subx.subsample()
```

```python
w = x[idx]
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2)

axs[0, 0].hist(x, 20, density=True, histtype='stepfilled', facecolor='g',
               alpha=0.75)
axs[0, 0].set_title('original')

axs[0, 1].hist(w, 20, density=True, histtype='stepfilled', facecolor='b',
               alpha=0.75)
axs[0, 1].set_title('subsampled')

axs[1, 0].hist(x, 20, density=False, histtype='stepfilled', facecolor='g',
               alpha=0.75)
axs[1, 0].set_title('original')

axs[1, 1].hist(w, 20, density=False, histtype='stepfilled', facecolor='b',
               alpha=0.75)
axs[1, 1].set_title('subsampled')
```
